```python
import pandas as pd
```

```python
daum_date = pd.DataFrame()
for _, row in esg.iterrows():
    if 'daum' in row['key']:
        date = row['key'][21:29]
        row['key'] = date[:4] + "." + date[4:6] + "." + date[6:]
        daum_date = pd.concat([daum_date, row], axis=1)
```

```python
daum_date = daum_date.T
```

```python
daum_date = daum_date.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 1)
```

```python
naver_date = pd.read_csv("../../data/ESG_split/naver_esg.csv")
```

```python
naver_date
```

```python
naver_date_org = pd.DataFrame()
for _, row in esg.iterrows():
    if 'naver' == row['portal']:
        naver_date_org = pd.concat([naver_date_org, row], axis=1)
```

```python
naver_date_org = naver_date_org.T
```

```python
naver_date['keyword'] = naver_date_org['keyword']
```

```python
naver_date = naver_date.drop(['Unnamed: 0'], axis = 1)
```

```python
naver_date.columns = ['url', 'title', 'body', 'date', 'keyword']
```

```python
naver_date
```

```python
daum_date.columns = ['title', 'body', 'date', 'portal', 'keyword']
daum_date
```

```python
naver_fin = naver_date[['title', 'body', 'date', 'keyword']]
```

```python
daum_fin = daum_date[['title', 'body', 'date', 'keyword']]
```

```python
news_date = pd.concat([naver_fin, daum_fin])
```

```python
for g in news_date.groupby('keyword'):
    temp = g[1]
    sample = temp.drop_duplicates('date')
    print(f"====={g[0]}======")
    for item in zip(sample['title'], sample['date']):
        print(item)
```
