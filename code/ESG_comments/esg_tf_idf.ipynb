```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
esg_comments = pd.read_csv("../data/ESG_split/comments_by_esg.csv")
```

```python
esg_comments
```

```python
document_lst = []
field = []
for item in esg_comments.groupby('esg'):
    field.append(item[0])
    document = ""
    for txt in pd.DataFrame(item[1])['text']:
        document = document + " " + txt 
    document_lst.append(document)
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
token_sent_lst = []
for item in document_lst:
    temp = okt.nouns(item)
    token_sent = ""
    for keyword in temp:
        token_sent = token_sent + " " + keyword
    token_sent_lst.append(token_sent)
```

```python
token_sent_lst
```

```python
vectorizer = CountVectorizer()
```

```python
dtm = vectorizer.fit_transform(token_sent_lst)
```

```python
dtm
```

```python
tf = pd.DataFrame(dtm.toarray(), columns = vectorizer.get_feature_names())
```

```python
tf
```

```python
df = tf.astype(bool).sum(axis = 0)
```

```python
# 문서 개수
D = len(tf)

# Inverse Document Frequency
idf = np.log((D+1) / (df+1)) + 1
```

```python
tfidf = tf * idf                      
tfidf = tfidf / np.linalg.norm(tfidf, axis = 1, keepdims = True)
```

```python
sort_tfidf_lst = []
for _, row in tfidf.iterrows():
    sort_tfidf_lst.append(list(row.sort_values(ascending=False).head(100).index))
```

```python
field
```

```python
tfidf_df = pd.DataFrame(sort_tfidf_lst).T
tfidf_df.columns = field
```

```python
tfidf_df.to_csv("../result/esg_comments_tfidf.csv")
```
