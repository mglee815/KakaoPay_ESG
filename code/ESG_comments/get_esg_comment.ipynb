```python
import pandas as pd
```

```python
naver_comments = pd.read_csv("../../data/KP_naver_comments_monthlyTop4000.csv")
daum_comments = pd.read_csv("../../data/KP_daum_comments_monthlyTop800.csv")
```

```python
naver_comment_UT = naver_comments[['url', 'text']]
daum_comment_UT = daum_comments[['org_url', 'content']]

daum_comment_UT.columns = ['url', 'text']
```

```python
news_comments = pd.concat([naver_comment_UT, daum_comment_UT])
```

```python
news_comments = news_comments.reset_index(drop = True)
```

```python
news_comments['url']
```

```python
news_comments['key'] = news_comments['url'].apply(lambda x : x[-22:-8] if 'naver' in x else x)
```

```python
news_comments  = news_comments.set_index('key')
```

```python
news_comments
```

```python
news = pd.read_csv("../../data/ESG_split/ESG_merge.csv")
```

```python
news[news['keyword'] == '자활기업 응원 프로젝트']
```

```python
news_comments
```

```python
comment_merge = pd.DataFrame()
for _, row in news.iterrows():
    idx = row['key']

    temp = news_comments.loc[idx,:].copy()
    temp = pd.DataFrame(temp)
    
    if temp.shape[1] == 1:
        temp = temp.T
    temp['esg'] = row['keyword']
    comment_merge = pd.concat([comment_merge, temp])
```

```python
comment_merge.groupby('esg').size()
```

```python
comment_merge = comment_merge.drop_duplicates()
```

```python
comment_merge.to_csv("../data/ESG_split/comments_by_esg.csv")
```
