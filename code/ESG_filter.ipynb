```python
TB_naver = pd.read_csv("../data/KP_naver_TB_0203.csv")
```

```python
TB_naver
```

```python
TB_naver = TB_naver.drop_duplicates(['title', 'body'])
```

```python
score_lst = []
for idx, item in TB_naver.iterrows():
    score = 0
    for keyword in ESG_keyword.keys():
        if keyword in item['body']:
            score += ESG_keyword[keyword]
    score_lst.append(score)
```

```python
TB_naver['score'] = score_lst.copy()
```
