```python
import pandas as pd
```

```python
df = pd.read_excel("../data/KP_2022.xlsx")
```

```python
df
```

```python
body = df.PTTN_CNTN_CL
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
tokens = body.apply(lambda x : okt.nouns(str(x)))
```

```python
from collections import Counter   
```

```python
tf = Counter()

for token in tokens:
    tf.update(token)
```

```python
pd.DataFrame(tf.most_common()).to_csv("../result/KP_TF.csv")
```
