```python
naver = pd.read_csv("../result/KP_naver_token.csv")
```

```python
from collections import Counter

naver_tf = Counter()

for token in naver['token']:
    naver_tf.update(eval(token))

naver_tokens_lst = list(zip(naver_tf.keys(), naver_tf.values()))

naver_tf_df = pd.DataFrame(naver_tokens_lst)
```

```python
naver_tf_filter = naver_tf_df[naver_tf_df[0].apply(lambda x : len(x) > 1)]
naver_tf_filter.sort_values(1, ascending=False, inplace=True)
```

```python
naver_tf_filter.columns = ['naver', 'n_freq']
naver_tf_filter.to_csv("../result/KP_naver_token_freq.csv")
```

```python
from knockknock import discord_sender

url = "https://discord.com/api/webhooks/989774347466047529/c5ZeC8lk3gX8Jq8i77WVuIByZhPyNbTwLPIaY-zQEiH9flKeNcWb6hyUODRUFO0sfnKm"
```

```python
stop_lst = [
    '안녕하세요 한글문화연대 연구원 신수호',
    '한글문화연대'
]
```

```python
def stop_lst_check(lst, x):
    for item in lst:
        if item in x:
            return False
    return True
```

```python
check = complain['PTTN_CNTN_CL'].apply(lambda x : stop_lst_check(stop_lst, str(x)))
```

```python
complain = complain[check]
```

```python
@discord_sender(webhook_url = url)
def tokenization(lst, x):
    okt = Okt()
    for item in x:
        lst.append(okt.nouns(str(item)))
    return {"done"}
```
