```python
w_dict = {}
for item in word_lst:
    item = item.strip()
    v, k = item.split('*')
    k = k.replace('"', "")
    v = v.replace('"', "")
    k = k.replace("'", "")
    v = v.replace("'", "")
    w_dict[k] = float(v)
```

```python
from wordcloud import WordCloud
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('./pay_logo.png')
img
img_array = np.array(img)

wc =WordCloud(font_path= "usr/share/fonts/truetype/nanum/NanumGothic.ttf", background_color="white",
              width=400, height=400, mask = img_array, colormap='Paired').generate_from_frequencies(w_dict)
```

```python
plt.figure()
plt.imshow(wc)
plt.axis('off')
```
