```python
from konlpy.tag import Okt

okt = Okt()
```

```python
tokens = youtube.comment.apply(lambda x : okt.nouns(str(x)))
```

```python
from collections import Counter   
tf = Counter()

for token in tokens:
    tf.update(token)
```

```python
tf_df = pd.DataFrame([tf]).T
```

```python
tf_df.sort_values(0, ascending=False).head(30)
```
